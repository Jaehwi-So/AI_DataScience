1. library import

```python
import matplotlib.pyplot as plt
import numpy as np
```

2. 함수 정의 y = (x-5)^2 - 20

```python
def f(x):
  return np.power(x-5, 2) - 20
  
```

3. 함수의 미분 정의 y' = 2(x-5)

```python
def f_derivative(x):
  return 2 * (x-5)
```

4. Gradient Descent 함수 정의

```python
def gradient_descent(x_curr, alpha, precision, max_iteration):  #시작점, 학습률, 임계치, 임계반복
  list_step = []  #x_curr와 x_next의 차이를 저장

  for i in range (max_iteration):
    x_next = x_curr - alpha * f_derivative(x_curr);
    step = x_next - x_curr;
    list_step.append (abs(step))  #절댓값 저장

    if(i % 100 == 0):
      print(i, x_next)

    if(abs(step) <= precision):
      break;

    x_curr = x_next

  print('min at x = ', x_next)

  #결과를 그림으로 보여주기
  figure, ax  = plt.subplots(1, 1)  #그림과 좌표측
  ax.title.set_text('step_size')
  ax.plot(list_step)
  ax.set_ylabel('step size')
  ax.set_xlabel('iter number')

  plt.show()

```

5. 결과 확인

```python
gradient_descent(x_curr = -10, alpha = 0.01, precision = 0.000001, max_iteration = 1000)
```

```python
gradient_descent(x_curr = 10, alpha = 0.01, precision = 0.0001, max_iteration = 1000)
```
